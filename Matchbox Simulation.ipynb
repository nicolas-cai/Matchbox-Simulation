```julia
#take integer n as input and return the factorial of n
function matchbox(a, b)
    if a == 0 && b == 0
        return 1
    elseif b ==0
        return 1 + 0.5*matchbox(a-1, 0)
    elseif(a==b)
        return 1 + matchbox(a, a-1)
    else
        return (1 + 0.5*(matchbox(a-1, b) + matchbox(a, b-1)))
    end
end
matchbox(100,100)
```

```julia

B = zeros(101,101)

B[1,1]=1.0

for c in 2:101

  B[c,1] = 1 + 0.5*B[c-1,1]

end

for c in 2:101

  for d in 2:c

    if d<c

      B[c,d] = 1 + 0.5*(B[c-1,d]+B[c,d-1])

    else

      B[c,d] = 1 + B[c,d-1]

    end

  end

end
                    
attempts = last(B)        
       
left = 200 - attempts + 1
left
```

```julia
function monteCarloTwo()
    X = 100
    Y = 100
    num = 0
    while X > -1 && Y > -1
        p = rand()
        if p > 0.5
            X -= 1
        else
            Y -= 1
        end
        num += 1
    end
    left = 200 - num + 1
    return left
end

sum = 0
for i in 1:1000
    sum += monteCarloTwo()
end

averageLeft = sum/1000
print(averageLeft)
```

```julia
y = []
for i in 1:1000
    append!(y, monteCarloTwo())
end

xlabel("Remaining Number of Matches")
ylabel("Frequency")
PyPlot.title("Matches Left W/ Two Boxes")
hist(y, 10)
```

```julia
#This is the recursive computation
B = zeros(101,101,101)

B[1,1,1]=1.0

for c in 2:101

  B[c,1,1] = 1 + 0.33*B[c-1,1]

end

for c in 2:101

  for d in 2:c

    if d<c

      B[c,d,1] = 1 + 0.33*(B[c-1,d,1]+B[c,d-1,1])

    else

      B[c,d,1] = 1 + 0.33*B[c,d-1,1]

    end
    

  end

end
                    
attempts = last(B)        
       
left = 200 - attempts + 1
left
```

```julia
function monteCarloThree()
    X = 100
    Y = 100
    Z = 100
    num = 0
    remain = 0
    while X > 0 && Y > 0 && Z > 0
        p = rand()
        if p > 0.66
            X -= 1
        elseif p > 0.33 && p < 0.66
            Y -= 1
        else 
            Z -= 1
        end
        num += 1
        if X == 0
            remain = Y
        elseif Y == 0
            remain = Z
        elseif Z == 0 
            remain = X
        end
    end
    left = 300 - num + 1
    leftFirst = left - remain
    return leftFirst
end

monteCarloThree()
```

```julia
sum = 0
for i in 1:1000
    sum += monteCarloThree()
end

averageLeft = sum/1000
print(averageLeft)
```
